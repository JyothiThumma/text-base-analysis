```
#Create pipeline, remove stop words, word frequency analysis and wordclouds.
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK stopwords
nltk.download('stopwords')
nltk.download('punkt')

# Define the anime dataset as a list of dictionaries
anime_data = [
    {'anime_id': 1, 'name': 'Cowboy Bebop', 'genre': 'Action', 'type': 'TV', 'episodes': 26, 'rating': 8.81, 'members': 893100},
    {'anime_id': 2, 'name': 'Cowboy Bebop: Tengoku no Tobira', 'genre': 'Action', 'type': 'Movie', 'episodes': 1, 'rating': 8.41, 'members': 116768},
    {'anime_id': 3, 'name': 'Cowboy Bebop: Yose Atsume Blues', 'genre': 'Action', 'type': 'Movie', 'episodes': 1, 'rating': 7.76, 'members': 14695},
    {'anime_id': 4, 'name': 'Trigun', 'genre': 'Action', 'type': 'TV', 'episodes': 26, 'rating': 8.31, 'members': 316481},
    {'anime_id': 5, 'name': "Wolf's Rain", 'genre': 'Action', 'type': 'TV', 'episodes': 26, 'rating': 7.92, 'members': 93351},
    {'anime_id': 6, 'name': "Wolf's Rain OVA", 'genre': 'Action', 'type': 'OVA', 'episodes': 4, 'rating': 7.53, 'members': 18221},
    {'anime_id': 7, 'name': 'Witch Hunter Robin', 'genre': 'Action', 'type': 'TV', 'episodes': 26, 'rating': 7.37, 'members': 54951},
    {'anime_id': 8, 'name': 'Bouken Ou Beet', 'genre': 'Action', 'type': 'TV', 'episodes': 52, 'rating': 6.98, 'members': 28539},
    {'anime_id': 9, 'name': 'Bouken Ou Beet Excellion', 'genre': 'Action', 'type': 'TV', 'episodes': 25, 'rating': 7.18, 'members': 25331},
    {'anime_id': 10, 'name': 'Eiyuu Gaiden Mozaicka', 'genre': 'Action', 'type': 'OVA', 'episodes': 1, 'rating': 7.16, 'members': 11236}
]

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(anime_data)

# Define stop words
stop_words = set(stopwords.words('english'))

# Define function to remove stop words
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_text = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

# Define function for word frequency analysis
def word_frequency(text):
    tokens = word_tokenize(text)
    freq_dist = nltk.FreqDist(tokens)
    return freq_dist

# Define function to generate word cloud
def generate_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Define pipeline
def text_processing_pipeline(text):
    text = remove_stopwords(text)
    freq_dist = word_frequency(text)
    generate_wordcloud(text)
    return freq_dist

# Apply pipeline to anime reviews
for index, row in df.iterrows():
    print("Anime:", row['name'])
    print("Rating:", row['rating'])
    freq_dist = text_processing_pipeline(row['genre'])
    print("Word Frequencies:", freq_dist)
    print("="*50)
```

```
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Define the health-related dataset
health_dataset = [
    "I recently started a new fitness routine and I feel great! My energy levels are up and I'm sleeping better.",
    "I've been struggling with chronic pain for years, and it's really affecting my quality of life. I'm desperate for relief.",
    "I just received some concerning test results from my doctor, and I'm feeling really anxious about what it could mean for my health.",
    "I've been trying to eat healthier and incorporate more fruits and vegetables into my diet, but it's been a challenge to stay consistent.",
    "I had a check-up with my doctor today, and I'm relieved to hear that my blood pressure and cholesterol levels are within the normal range.",
    "I've been feeling really fatigued lately, and I'm worried that it could be a sign of a more serious health issue.",
    "I'm proud of myself for quitting smoking and making positive changes to improve my lung health.",
    "I injured myself while exercising, and now I'm worried that I won't be able to continue with my fitness routine.",
    "I've been taking medication for my condition, and I'm happy to report that I'm seeing some improvement in my symptoms.",
    "I'm feeling optimistic about my health goals for the new year and excited to make progress towards them."
]

# Initialize the sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each sentence in the health-related dataset
for i, sentence in enumerate(health_dataset):
    # Analyze the sentiment of the sentence
    sentiment_scores = sid.polarity_scores(sentence)

    # Determine the sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        sentiment_label = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    # Print the sentiment analysis results
    print(f"Sentence {i+1}: {sentence}")
    print(f"Sentiment Score: {sentiment_scores['compound']}")
    print(f"Sentiment Label: {sentiment_label}")
    print("=" * 50)
```

```
import re
import matplotlib.pyplot as plt

# Define updated example strings
Example_bharat = [
    "joythi2854@mail.yahoo.com",
    "jul",
    "2024-02-14",
    "192.168.1.1"
]

# Define regex patterns for each example type
regex_patterns = [
    r'^[\w\.-]+@[\w\.-]+\.\w+$', # Email address
    r'^[a-zA-Z_][a-zA-Z0-9_]*$',  # Variable name in Python
    r'^\d{4}-\d{2}-\d{2}$',        # Date in YYYY-MM-DD format
    r'^((25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)$', # IP address (IPv4)
    r'^(https?://)?(www\.)?([\w-]+\.[\w-]+)(/[\w- ./?%&=]*)?$' # URL
]

# Initialize counters for matched and not matched strings
matched_counts = [0] * len(regex_patterns)
not_matched_counts = [0] * len(regex_patterns)

# Test regex patterns on updated example strings
for i, string in enumerate(Example_bharat):
    matched = re.match(regex_patterns[i], string)
    if matched:
        matched_counts[i] += 1
    else:
        not_matched_counts[i] += 1

# Plot the results
fig, ax = plt.subplots(figsize=(10, 6))
patterns = ['Email', 'Variable Name', 'Date', 'IP Address', 'URL']
x = range(len(patterns))

ax.bar(x, matched_counts, width=0.4, label='Matched', color='blue')
ax.bar(x, not_matched_counts, width=0.4, bottom=matched_counts, label='Not Matched', color='red')

ax.set_xlabel('Regex Patterns')
ax.set_ylabel('Count')
ax.set_title('Matching Results of Regex Patterns')
ax.set_xticks(x)
ax.set_xticklabels(patterns, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
